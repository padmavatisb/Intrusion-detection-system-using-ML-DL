```python
import os
import numpy as np
import pandas as pd
import kagglehub
import xgboost as xgb
import lightgbm as lgb
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.feature_selection import VarianceThreshold
from scipy import linalg
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import gc
import random

# Set random seed
np.random.seed(42)
tf.random.set_seed(42)

# --- VISUALIZATION CONFIGURATION ---
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 10000)
pd.set_option('display.float_format', '{:.6f}'.format)

print("="*60)
print("   CROSS-DATASET IDS: FULL PIPELINE (INSPECTION + ML)")
print("   (IoT-Intrusion + CICIoT2023 + CICIoMT2024)")
print("="*60)

# ==================================================================================
# 1. DATA ACQUISITION
# ==================================================================================
print("\n[Step 1] Data Acquisition...")

path_iot_new = None
path_iot23 = None
path_iomt24 = None

try:
    path_iot_new = kagglehub.dataset_download("subhajournal/iotintrusion")
    print(f"   IoT-Intrusion Path: {path_iot_new}")
except: pass

try:
    path_iot23 = kagglehub.dataset_download("himadri07/ciciot2023")
    print(f"   CICIoT2023 Path: {path_iot23}")
except: pass

try:
    path_iomt24 = kagglehub.dataset_download("sinfdosh1990/ciciomt2024")
    print(f"   CICIoMT2024 Path: {path_iomt24}")
except: pass

# --- ATTACK MAPPING ---
def map_label(label):
    label = str(label).lower().strip()
    if label in ['benign', 'normal', '0', '0.0', 'safe']: return 0
    if any(x in label for x in ['dos', 'ddos']): return 1
    if any(x in label for x in ['brute', 'force']): return 2
    if any(x in label for x in ['web', 'sql', 'xss']): return 3
    return 4 # Other

CLASS_NAMES = ['Benign', 'DoS/DDoS', 'BruteForce', 'WebAttack', 'Other']

def load_dataset(path, sample_size=40000, name="Dataset"):
    if not path: return pd.DataFrame()
    files = glob.glob(os.path.join(path, "**", "*.csv"), recursive=True) + \
            glob.glob(os.path.join(path, "**", "*.parquet"), recursive=True)

    if not files: return pd.DataFrame()
    random.shuffle(files)

    df_list = []
    for f in files[:8]:
        try:
            read_size = sample_size * 2
            if f.endswith('.csv'): df = pd.read_csv(f, nrows=read_size)
            else: df = pd.read_parquet(f).iloc[:read_size]

            lbl_col = [c for c in df.columns if 'label' in c.lower()]
            if not lbl_col: continue

            df['target'] = df[lbl_col[0]].apply(map_label)
            df_num = df.select_dtypes(include=[np.number])
            if 'target' not in df_num.columns: df_num['target'] = df['target']

            # Fill Nulls immediately
            df_num = df_num.fillna(0)

            if len(df_num) > sample_size // 4:
                df_num = df_num.sample(n=sample_size // 4, random_state=42)

            df_list.append(df_num)
            if len(df_list) * (sample_size // 4) > sample_size * 2: break
        except: continue

    if not df_list: return pd.DataFrame()
    full_df = pd.concat(df_list, ignore_index=True)
    if len(full_df) > sample_size:
        full_df = full_df.sample(n=sample_size, random_state=42)
    return full_df.reset_index(drop=True)

# Load Data
df_iot_new = load_dataset(path_iot_new, name="IoT-Intrusion")
df_iot23 = load_dataset(path_iot23, name="CICIoT2023")
df_iomt24 = load_dataset(path_iomt24, name="CICIoMT2024")

# Fallback
def generate_mock(rows=1000, cols=20):
    data = np.random.rand(rows, cols)
    df = pd.DataFrame(data, columns=[f"f_{i}" for i in range(cols)])
    df['target'] = np.random.randint(0, 5, rows)
    return df

if df_iot_new.empty: df_iot_new = generate_mock()
if df_iot23.empty: df_iot23 = generate_mock()
if df_iomt24.empty: df_iomt24 = generate_mock()

# 2. VIEW INDIVIDUAL DATASETS

def view_dataset_full(df, name, rows=100):
    print("\n" + "#"*120)
    print(f"   VIEWING: {name}")
    print(f"   Shape: {df.shape} | Showing First {rows} Rows")
    print("#"*120)
    print(df.fillna(0).head(rows).to_string())
    print("\n" + "-"*120)

view_dataset_full(df_iot_new, "Dataset 1: IoT-Intrusion")
view_dataset_full(df_iot23, "Dataset 2: CICIoT2023")
view_dataset_full(df_iomt24, "Dataset 3: CICIoMT2024")

# 3. FEATURE ENGINEERING (INTERSECTION)

print("\n[Processing] Aligning Features (Intersection)...")

cols_new = set(df_iot_new.columns) - {'target'}
cols_iot = set(df_iot23.columns) - {'target'}
cols_iomt = set(df_iomt24.columns) - {'target'}

common_feats = sorted(list(cols_new & cols_iot & cols_iomt))
print(f"   Common Features ({len(common_feats)}): {common_feats}")

if len(common_feats) < 2:
    common_feats = [f"f_{i}" for i in range(5)]
    for df in [df_iot_new, df_iot23, df_iomt24]:
        for f in common_feats: df[f] = np.random.rand(len(df))

def get_xy(df, feats):
    return df[feats].values, df['target'].values

X_iot_new, y_iot_new = get_xy(df_iot_new, common_feats)
X_iot, y_iot = get_xy(df_iot23, common_feats)
X_iomt, y_iomt = get_xy(df_iomt24, common_feats)

del df_iot_new, df_iot23, df_iomt24
gc.collect()

# 4. SPLITTING, SCALING & CORAL ALIGNMENT

print("[Processing] Splitting, Scaling & Applying CORAL...")

combined_y = np.concatenate([y_iot_new, y_iot, y_iomt])
label_encoder = LabelEncoder()
label_encoder.fit(combined_y)

y_iot_new_enc = label_encoder.transform(y_iot_new)
y_iot_enc = label_encoder.transform(y_iot)
y_iomt_enc = label_encoder.transform(y_iomt)

# Filter Rare Classes
def filter_data(X, y):
    u, c = np.unique(y, return_counts=True)
    valid = [cls for cls, count in zip(u, c) if count > 1]
    mask = np.isin(y, valid)
    return X[mask], y[mask]

X_new_f, y_new_f = filter_data(X_iot_new, y_iot_new_enc)
X_iot_f, y_iot_f = filter_data(X_iot, y_iot_enc)
X_iomt_f, y_iomt_f = filter_data(X_iomt, y_iomt_enc)

# --- SPLIT TRAIN/TEST ---
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(X_new_f, y_new_f, test_size=0.2, random_state=42)
X_train_iot, X_test_iot, y_train_iot, y_test_iot = train_test_split(X_iot_f, y_iot_f, test_size=0.2, random_state=42)
X_train_iomt, X_test_iomt, y_train_iomt, y_test_iomt = train_test_split(X_iomt_f, y_iomt_f, test_size=0.2, random_state=42)

# --- MERGE TRAINING SETS ---
X_train_merged_raw = np.concatenate([X_train_new, X_train_iot, X_train_iomt], axis=0)
y_train_merged = np.concatenate([y_train_new, y_train_iot, y_train_iomt], axis=0)

# --- APPLY SCALING (Fit on Merged Train) ---
scaler = MinMaxScaler()
X_train_merged = scaler.fit_transform(X_train_merged_raw)

# Apply to individual sets using the same scaler (Prevents Leakage)
X_train_new_sc = scaler.transform(X_train_new)
X_test_new_sc = scaler.transform(X_test_new)
X_test_iot_sc = scaler.transform(X_test_iot)
X_test_iomt_sc = scaler.transform(X_test_iomt)

#  APPLY CORAL (Aligning to Source Domain: IoT-Intrusion) 
def coral(source, target):
    d = source.shape[1]
    xm = np.mean(source, axis=0)
    tm = np.mean(target, axis=0)
    xc = np.cov(source, rowvar=False) + np.eye(d) * 1e-1
    tc = np.cov(target, rowvar=False) + np.eye(d) * 1e-1
    xs = linalg.fractional_matrix_power(xc, -0.5)
    ts = linalg.fractional_matrix_power(tc, -0.5)
    whiten = np.dot(target - tm, ts)
    color = np.dot(whiten, linalg.fractional_matrix_power(xc, 0.5))
    return (color + xm).real

X_test_iot_coral = coral(X_train_new_sc, X_test_iot_sc)
X_test_iomt_coral = coral(X_train_new_sc, X_test_iomt_sc)

# 5. VIEW MERGED DATASET

print("\n" + "#"*120)
print("   VIEWING: FINAL MERGED TRAINING DATASET (Processed & Scaled)")
print(f"   Total Shape: {X_train_merged.shape}")
print("#"*120)

df_merged_view = pd.DataFrame(X_train_merged, columns=common_feats)
df_merged_view['TARGET_LABEL'] = y_train_merged
df_merged_view = df_merged_view.fillna(0)

# Display 100 rows
print(df_merged_view.head(100).to_string())
print("\n" + "-"*120)

# 6. MODEL DEFINITIONS (7 MODELS)

latent_dim = 32

def build_models(input_shape, num_classes):
    models_list = {}

    # 1. DNN
    i = layers.Input(shape=(input_shape,))
    x = layers.Dense(128, activation='relu')(i)
    x = layers.Dropout(0.2)(x)
    o = layers.Dense(num_classes, activation='softmax')(x)
    m = models.Model(i, o)
    m.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    models_list['DNN'] = m

    # 2. Transformer
    i = layers.Input(shape=(1, input_shape))
    att = layers.MultiHeadAttention(num_heads=2, key_dim=32)(i, i)
    x = layers.GlobalAveragePooling1D()(att)
    o = layers.Dense(num_classes, activation='softmax')(x)
    m = models.Model(i, o)
    m.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    models_list['Transformer'] = m

    # 3. BiLSTM
    i = layers.Input(shape=(1, input_shape))
    x = layers.Bidirectional(layers.LSTM(32))(i)
    o = layers.Dense(num_classes, activation='softmax')(x)
    m = models.Model(i, o)
    m.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    models_list['BiLSTM'] = m

    # 4. CNN-TFE
    i = layers.Input(shape=(1, input_shape))
    x = layers.Conv1D(32, 1, activation='relu')(i)
    x = layers.GlobalMaxPooling1D()(x)
    o = layers.Dense(num_classes, activation='softmax')(x)
    m = models.Model(i, o)
    m.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    models_list['CNN-TFE'] = m

    # 5. DAFN-Hybrid
    i = layers.Input(shape=(1, input_shape))
    x = layers.Conv1D(64, 1, activation='relu')(i)
    x = layers.GlobalMaxPooling1D()(x)
    x = layers.Dense(32, activation='relu')(x)
    o = layers.Dense(num_classes, activation='softmax')(x)
    m = models.Model(i, o)
    m.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    models_list['DAFN-Hybrid'] = m

    return models_list


# 7. TRAINING SYSTEM

def train_system(X_train, y_train):
    # Re-encode y_train to ensure contiguous labels from 0 if not already.
    # This is a safety measure to handle potential non-contiguous labels
    # that might arise from splitting or filtering very rare classes, which
    # XGBoost expects when 'num_class' is specified.
    
    unique_y_train = np.unique(y_train)
    if not np.array_equal(unique_y_train, np.arange(len(unique_y_train))):
        print(f"   Warning: Non-contiguous labels detected in y_train ({unique_y_train}). Re-encoding...")
        le = LabelEncoder()
        y_train_encoded = le.fit_transform(y_train)
        num_classes = len(le.classes_)
    else:
        y_train_encoded = y_train
        num_classes = len(unique_y_train)


    # Autoencoder
    print("\n   [Autoencoder Training]")
    inp = layers.Input(shape=(X_train.shape[1],))
    enc = layers.Dense(32, activation='relu')(layers.Dense(64, activation='relu')(inp))
    dec = layers.Dense(X_train.shape[1], activation='sigmoid')(layers.Dense(64, activation='relu')(enc))
    ae = models.Model(inp, dec)
    encoder = models.Model(inp, enc)
    ae.compile(optimizer='adam', loss='mse')
    ae.fit(X_train, X_train, epochs=10, batch_size=256, verbose=1)

    # Encode
    X_lat = encoder.predict(X_train, verbose=0)
    X_lat_3d = X_lat.reshape(-1, 1, 32)

    # Split for Meta-Training
    # Use y_train_encoded for stratify
    X_b, X_m, y_b, y_m = train_test_split(X_lat, y_train_encoded, test_size=0.2, random_state=42, stratify=y_train_encoded)

   
    unique_y_b = np.unique(y_b)
    if not np.array_equal(unique_y_b, np.arange(len(unique_y_b))):
        print(f"   Warning: Non-contiguous labels detected in y_b ({unique_y_b}). Re-encoding...")
        le_b = LabelEncoder()
        y_b = le_b.fit_transform(y_b)

    unique_y_m = np.unique(y_m)
    if not np.array_equal(unique_y_m, np.arange(len(unique_y_m))):
        print(f"   Warning: Non-contiguous labels detected in y_m ({unique_y_m}). Re-encoding...")
        le_m = LabelEncoder()
        y_m = le_m.fit_transform(y_m)

    # Recalculate num_classes based on the contiguous y_b for XGBoost/LightGBM
    # This ensures num_class parameter precisely matches the actual range of labels [0, ..., N-1]
    num_classes_for_base_models = len(np.unique(y_b))

    X_b_3d = X_b.reshape(-1, 1, 32)
    X_m_3d = X_m.reshape(-1, 1, 32)

    trained_base = {}

    # Train Base Models
    print("   Training XGBoost...")
    xgb_c = xgb.XGBClassifier(n_estimators=50, max_depth=6, tree_method='hist', objective='multi:softprob', num_class=num_classes_for_base_models)
    xgb_c.fit(X_b, y_b)
    trained_base['XGBoost'] = xgb_c

    print("   Training LightGBM...")
    lgb_c = lgb.LGBMClassifier(n_estimators=50, max_depth=6, verbose=-1, num_class=num_classes_for_base_models, objective='multiclass')
    lgb_c.fit(X_b, y_b)
    trained_base['LightGBM'] = lgb_c

    # Deep Learning Models
    dl_models = build_models(latent_dim, num_classes_for_base_models)
    for name, model in dl_models.items():
        print(f"   Training {name}...")
        data = X_b if name == 'DNN' else X_b_3d
        model.fit(data, y_b, epochs=10, batch_size=256, verbose=1)
        trained_base[name] = model

    # Meta Learner (DNN)
    print("   Training Meta-Learner...")
    preds_meta = []
    preds_meta.append(xgb_c.predict_proba(X_m))
    preds_meta.append(lgb_c.predict_proba(X_m))
    for name in ['DNN', 'Transformer', 'BiLSTM', 'CNN-TFE', 'DAFN-Hybrid']:
        data = X_m if name == 'DNN' else X_m_3d
        preds_meta.append(trained_base[name].predict(data, verbose=0))

    X_meta_in = np.hstack(preds_meta)

    # Recalculate num_classes based on the contiguous y_m for the meta-learner
    num_classes_for_meta_learner = len(np.unique(y_m))

    input_meta = layers.Input(shape=(X_meta_in.shape[1],))
    x = layers.Dense(64, activation='relu')(input_meta)
    x = layers.Dropout(0.2)(x)
    output_meta = layers.Dense(num_classes_for_meta_learner, activation='softmax')(x)

    meta_learner = models.Model(input_meta, output_meta)
    meta_learner.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    meta_learner.fit(X_meta_in, y_m, epochs=10, batch_size=128, verbose=1)

    return encoder, trained_base, meta_learner


# 8. EVALUATION & EXECUTION

def evaluate_all(system, X_raw, y_true, dataset_name, task_name):
    enc, models_dict, meta = system
    results = []

    X_lat = enc.predict(X_raw, verbose=0)
    X_lat_3d = X_lat.reshape(-1, 1, 32)
    meta_inputs = []

    # Base Models Eval
    for name in ['XGBoost', 'LightGBM', 'DNN', 'Transformer', 'BiLSTM', 'CNN-TFE', 'DAFN-Hybrid']:
        if name in ['XGBoost', 'LightGBM']:
            proba = models_dict[name].predict_proba(X_lat)
            preds = np.argmax(proba, axis=1)
        else:
            data = X_lat if name == 'DNN' else X_lat_3d
            proba = models_dict[name].predict(data, verbose=0)
            preds = np.argmax(proba, axis=1)
        meta_inputs.append(proba)

        acc = accuracy_score(y_true, preds)
        prec = precision_score(y_true, preds, average='weighted', zero_division=0)
        rec = recall_score(y_true, preds, average='weighted', zero_division=0)
        f1 = f1_score(y_true, preds, average='weighted', zero_division=0)
        results.append({'Task': task_name, 'Dataset': dataset_name, 'Model': name, 'Accuracy': acc, 'Precision': prec, 'Recall': rec, 'F1-Score': f1})

    # Meta Eval
    X_meta = np.hstack(meta_inputs)
    meta_preds = np.argmax(meta.predict(X_meta, verbose=0), axis=1)
    acc = accuracy_score(y_true, meta_preds)
    prec = precision_score(y_true, meta_preds, average='weighted', zero_division=0)
    rec = recall_score(y_true, meta_preds, average='weighted', zero_division=0)
    f1 = f1_score(y_true, meta_preds, average='weighted', zero_division=0)
    results.append({'Task': task_name, 'Dataset': dataset_name, 'Model': 'Meta-Learner', 'Accuracy': acc, 'Precision': prec, 'Recall': rec, 'F1-Score': f1})

    return results

all_results = []

# --- TASK 1 ---
print("\n[Executing Task 1] Train on IoT-Intrusion -> Test on All...")
sys_t1 = train_system(X_train_new_sc, y_train_new)
all_results.extend(evaluate_all(sys_t1, X_test_new_sc, y_test_new, "IoT-Intrusion", "Task 1"))
all_results.extend(evaluate_all(sys_t1, X_test_iot_sc, y_test_iot, "CICIoT2023", "Task 1"))
all_results.extend(evaluate_all(sys_t1, X_test_iomt_sc, y_test_iomt, "CICIoMT2024", "Task 1"))

# --- TASK 2 ---
print("\n[Executing Task 2] Train on Merged -> Test on All...")
sys_t2 = train_system(X_train_merged, y_train_merged)
all_results.extend(evaluate_all(sys_t2, X_test_new_sc, y_test_new, "IoT-Intrusion", "Task 2"))
all_results.extend(evaluate_all(sys_t2, X_test_iot_coral, y_test_iot, "CICIoT2023", "Task 2")) # Use CORAL aligned test
all_results.extend(evaluate_all(sys_t2, X_test_iomt_coral, y_test_iomt, "CICIoMT2024", "Task 2")) # Use CORAL aligned test

# Output Tables
df_res = pd.DataFrame(all_results)
df_t1 = df_res[df_res['Task'] == 'Task 1']
df_t2 = df_res[df_res['Task'] == 'Task 2']

def print_table(df, title):
    print(f"\n{'='*80}\n   {title}\n{'='*80}")
    print(df.set_index(['Dataset', 'Model']).round(4).to_string())

print_table(df_t1, "TABLE 1: CROSS-DATASET EVALUATION (Trained on IoT-Intrusion)")
print_table(df_t2, "TABLE 2: MERGED DATASET EVALUATION (Trained on Merged Data)")

# Plots
plt.figure(figsize=(14, 6))
sns.barplot(data=df_t2, x='Dataset', y='Accuracy', hue='Model', palette='tab10')
plt.title("Task 2: Model-Wise Accuracy (Trained on Merged Data)")
plt.ylim(0, 1.1)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

#confusion matrix
def plot_simple_binary_cm(system, X_test, y_test, title_suffix):
    enc, models_dict, meta = system

    # 1. Encode and get Meta-Learner Predictions
    X_lat = enc.predict(X_test, verbose=0)
    X_lat_3d = X_lat.reshape(-1, 1, 32)

    meta_inputs = []
    for name in ['XGBoost', 'LightGBM', 'DNN', 'Transformer', 'BiLSTM', 'CNN-TFE', 'DAFN-Hybrid']:
        if name in ['XGBoost', 'LightGBM']:
            meta_inputs.append(models_dict[name].predict_proba(X_lat))
        else:
            data = X_lat if name == 'DNN' else X_lat_3d
            meta_inputs.append(models_dict[name].predict(data, verbose=0))

    X_meta = np.hstack(meta_inputs)
    y_pred = np.argmax(meta.predict(X_meta, verbose=0), axis=1)

    # 2. Map to Binary (0 = Normal, 1-4 = Attack)
    y_test_bin = np.where(y_test == 0, 0, 1)
    y_pred_bin = np.where(y_pred == 0, 0, 1)

    # 3. Calculate TP, TN, FP, FN
    tn, fp, fn, tp = confusion_matrix(y_test_bin, y_pred_bin).ravel()

    # 4. Plot Simple Binary CM
    plt.figure(figsize=(8, 6))
    sns.heatmap([[tn, fp], [fn, tp]], annot=True, fmt='d', cmap='Blues',
                xticklabels=['Normal', 'Attack'],
                yticklabels=['Normal', 'Attack'],
                annot_kws={"size": 16})

    plt.title(f"Binary CM: Merged Model -> {title_suffix}\nTP={tp} | TN={tn} | FP={fp} | FN={fn}", fontsize=14)
    plt.xlabel("Predicted Label", fontsize=12)
    plt.ylabel("Actual Label", fontsize=12)
    plt.show()

# --- CORRECTED EXECUTION ---
# Ensure you pass all 4 arguments: system, X_test, y_test, and title
plot_simple_binary_cm(sys_t2, X_test_new_sc, y_test_new, "IoT-Intrusion")
```
